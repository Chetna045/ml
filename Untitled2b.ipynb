```python
from sklearn.svm import SVC
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/heart.csv")
df
```

```python
x=df.loc[:,df.columns!='target']
x
```

```python
y=df['target'].values
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# we need to do fit with linear first
linear=svm.SVC(kernel='linear')
linear.fit(x_train,y_train)
```

```python
y_pred=linear.predict(x_test)
```

```python
print("Accuracy is",accuracy_score(y_test,y_pred))
print("confusion matrix is",confusion_matrix(y_test,y_pred))
print("classification report is",classification_report(y_test,y_pred))
```

```python
# we need to do for other models as well ->sigmoid,poly and rbf
model1=svm.SVC(kernel='sigmoid')
model2=svm.SVC(kernel='poly')
model3=svm.SVC(kernel='rbf')
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)

```

```python
y_pred1=model1.predict(x_test)
y_pred2=model2.predict(x_test)
y_pred3=model3.predict(x_test)
```

```python
print("Accuracy  for sigmoid model is",accuracy_score(y_test,y_pred1))
print("Accuracy for poly model is",accuracy_score(y_test,y_pred2))
print("Accuracy for rbf model is",accuracy_score(y_test,y_pred3))
```
